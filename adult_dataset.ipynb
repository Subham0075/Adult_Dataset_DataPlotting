```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = pd.read_csv(r"/content/adult.csv")
a
```

```python
a.dtypes
```

```python
a.isnull().sum()
```

```python
for i in a.columns:
    print(i,':','\n',a[i].unique())
```

```python
for i in a.columns:
    print(i,':','\t',sum(a[i]=='?'))
```

```python
a.workclass.describe()
```

```python
a.workclass = a.workclass.replace('?','Private')
```

```python
a.workclass.unique()
```

```python
a.occupation.describe()
```

```python
a.workclass = a.workclass.replace('?','Prof-specialty')
a.workclass.unique()
```

```python
a['native-country'].describe()
```

```python
a.size
```

```python
a.shape
```

```python
a['native-country'] = a['native-country'].replace('?','United-States')
a['native-country'].unique()
```

```python
#countplot
sns.countplot(x=a.income,hue=a.gender)
plt.show()
```

```python
#box plot - categorical+numerical+hue=(0,1)
sns.boxplot(x=a.gender,y=a.age,hue=a.race)
plt.show()
```

```python
sns.boxplot(x=a.gender,y=a.age,hue=a.income)
plt.show()
```

```python
sns.kdeplot(x=a.age)
plt.show()
```

```python
sns.displot(x=a.age)
plt.show()
```

```python
sns.displot(a.income[a.gender=='Male'],label='M')
sns.displot(a.income[a.gender=='Female'],label='F')
plt.legend()
plt.show()
```

```python
#violin plot
sns.violinplot(x=a.age,y=a.income,hue=a.gender)
plt.show()
```

```python
#pair plot
sns.pairplot(a,hue='income')
plt.show()
```

```python
#heatmap
cor = a.corr()
cor
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(cor,annot=True,cmap='coolwarm')
plt.show()
```
